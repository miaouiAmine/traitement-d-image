```python
# Import des bibliothques ncessaires
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Charger l'image
image_path = 'image1.jpg'  # Remplace par le chemin de ton image
image1 = Image.open(image_path)

# Redimensionner l'image
image_resized = image1.resize((256, 256))  # Redimensionner  une taille de 256x256

# Convertir l'image en un tableau numpy pour appliquer des convolutions
image_array = np.array(image_resized)

# Convertir l'image en niveaux de gris (si ce n'est pas dj le cas)
gray_image = cv2.cvtColor(image_array, cv2.COLOR_RGB2GRAY)

# Appliquer plusieurs filtres pour l'extraction de caractristiques

# Filtre Sobel (Dtection des bords horizontaux et verticaux)
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)  # Bords horizontaux
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)  # Bords verticaux
sobel_edge = cv2.magnitude(sobel_x, sobel_y)  # Magnitude des bords

# Filtre Laplacien (Dtection des bords)
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)

# Filtre de Gaussien (pour lisser l'image et dtecter les zones homognes)
gaussian_blur = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Affichage des rsultats
plt.figure(figsize=(12, 12))

# Afficher l'image originale redimensionne
plt.subplot(2, 3, 1)
plt.imshow(image_resized)
plt.title("Image Redimensionne")
plt.axis('off')

# Afficher l'image en niveaux de gris
plt.subplot(2, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Image en Niveaux de Gris")
plt.axis('off')

# Afficher le rsultat du filtre Sobel (bords)
plt.subplot(2, 3, 3)
plt.imshow(sobel_edge, cmap='gray')
plt.title("Dtection de Bords (Sobel)")
plt.axis('off')

# Afficher le rsultat du filtre Laplacien
plt.subplot(2, 3, 4)
plt.imshow(laplacian, cmap='gray')
plt.title("Dtection de Bords (Laplacien)")
plt.axis('off')

# Afficher le rsultat du flou Gaussien
plt.subplot(2, 3, 5)
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Flou Gaussien")
plt.axis('off')

# Montrer l'image du rsultat de chaque filtre
plt.tight_layout()
plt.show()
```
